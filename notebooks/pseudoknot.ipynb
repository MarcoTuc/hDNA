```python
def wc(a):
    wcd = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    return ''.join([wcd[i] for i in a])

def wcwhere(a,b):
    wcd = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    return [a[i] == wcd[b[i]] for i in range(len(a))]

def addparwhere(string, i, where, char):
    for j, wc in enumerate(where):
        if wc:
            string = string[:i+j] + char + string[i+j+1:]
    return string

e1 = 'GGTC'
e2 = 'CGAG'
print(wcwhere(e1, e2))

addparwhere('....', 0, wcwhere(e1,e2), '(')
```
