```python
d2 = 3e13
d3 = 7e9

px.scatter(np.geomspace(3e13,7e9,1000))
```

```python
def geospace(start,stop,num):
    ratio = (stop / start) ** (1 / (num - 1))
    return np.array([start * ratio ** (i) for i in range(num)])
px.scatter(geospace(d2,d3,1000))
```

```python
def my_transformed_geomspace(start, stop, num):
    # Generate a "standard" geometric sequence
    seq = np.geomspace(start, stop, num)
    # Apply the transformation function
    seq = seq
    # Scale and shift to match the desired start and stop values
    scale_factor = (stop - start) / (seq[-1] - seq[0])
    shifted_seq = (seq - seq[0]) * scale_factor + start
    return shifted_seq
px.scatter(my_transformed_geomspace(d2,d3,12))
```
